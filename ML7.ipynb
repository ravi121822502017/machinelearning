### Linear Regression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
help(make_regression)
```

```python
X,y = make_regression(100,1,n_targets=1,random_state=34)
X
```

```python
np.max(X)
```

```python
np.min(y)
```

```python
print(X.shape)
print(type(X))
print(X.ndim)
print(y.shape)
print(type(y))
print(y.ndim)
```

```python
plt.plot(X,y,'o')
```

```python
# import the Linear regression model
from sklearn.linear_model import LinearRegression
# Create an instance
LR = LinearRegression()
# Train
LR.fit(X,y)
# Predict with train data
y_pred = LR.predict(X)
```

```python
print(LR.coef_)
print(LR.intercept_)
```
