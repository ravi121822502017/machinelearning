```python
# import the Linear Regression Model
from sklearn.linear_model import LinearRegression
# Create an instance
LR=LinearRegression()
# Train
LR.fit(X,y)
# Predict with train data
y_pred=LR.predict(X)
```

```python
# Create user defined function with name mse
# inputs y_pred, y
# output is mean squared error
def mse(y_pred,y):
    return (y_pred- y)**2).mean()
```
